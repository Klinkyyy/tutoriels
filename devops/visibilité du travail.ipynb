# Visibilité du travail au sein de l'équipe et de l'entreprise

## Pourquoi ?

### Il n’y a pas que le développement qui compte…

Le développement logiciel ne fonctionne pas seul dans le vide. 

Il est une des activités qui permettent à une organisation (entreprise, association, …) de mettre à disposition de ses clients un produit qui leur apporte de la valeur.

Une des erreurs que l’on retrouve souvent en entrerprise, c’est de chercher à sur-optimiser l’étape de développement logiciel alors que ce sont d’autres étapes de la chaîne de valeur qui posent problème.

### Qu’est-ce qu’une chaîne de valeur ?

C’est l’ensemble des étapes et processus qui mènent d’une idée à un produit utilisable par un client.

- Cela inclut toutes les étapes, de l'idée jusqu'au déploiement.
- Il s'agit d'optimiser la chaîne dans son ensemble, pas seulement le développement.

## Comment donner de la visibilité ?

### Cartographie de la chaîne de valeur

- **temps d’écoulement (lead time)** - le temps qui s’écoule entre le moment où un processus accepte une unité de travail, et celui où il transfère cette unité au processus aval.
- **temps de traitement (process time)** - le temps que prendrait un processus s’il était exécuté par une personne seule (sans blocage).
- **fiabilité** - le pourcentage de fois où un processus fournit un résultat pertinent et utilisable sans retraitement.

### Management visuel

Permettre en un coup d’oeil de savoir quelles sont les unités de travail en cours, lesquelles sont bloquées, lesquelles sont terminées, lesquelles on doit commencer ensuite.

Avoir une idée claire des problèmes et urgences (incident de production, …).
Le tout dans le but d’éviter de nécessiter de grandes quantités de communications pour que tout le monde ait le même niveau d’information.

**Exemples de management visuel**

- Tableaux Kanban
- Diagrammes de flux cumulés
- Outil de monitoring des systèmes en production
- Outil de monitoring du nombre d’appels client / de commandes etc.

Affichés dans le bureau ou dans un onglet du navigateur, ils ne concernent pas forcément que l'équipe.
